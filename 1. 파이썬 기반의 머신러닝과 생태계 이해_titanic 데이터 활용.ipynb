#### dataframe 데이터 삭제

```python
dataframe.drop(labels = None, axis = 0, index = None, columns = None, level = None, inplace = False, errors = 'raise')
```

- axis = 0이면 행 삭제, axis = 1이면 칼럼 삭제
- labels에 원하는 칼럼명을 입력하고 axis = 1을 입력하면 지정된 칼럼을 삭제
- 여러 개의 칼럼믕ㄹ 삭제하려면 labels에 칼럼명을 리스트 형태로 작성
- inplace = False가 디폴트. inplace = True로 설정해야 원본 데이터가 변함

#### dataframe iloc, loc연산자

- iloc: 위치 기반 인덱싱 방식: 행과 열 위치를, 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값으로 지정하는 방식
- loc: 명칭 기반 인덱싱 방식: 데이터 프레임의 인덱스 값으로 행 위치를, 칼럼의 명칭으로 열 위치를 지정하는 방식

````python
data = {'Name' : ['chulmin', 'eunkyung', 'jinwoong', 'soobeom'],
       'Year' : [2011, 2016, 2015, 2015],
       'Gender' : ['male', 'female', 'male', 'male']}
data_df = pd.DataFrame(data, index = ['one', 'two', 'three', 'four'])
data_df
````

````python
print(data_df.iloc[0, 0])
print(data_df.iloc[1, 0])
print(data_df.iloc[2, 1])
print(data_df.iloc[0:2, [0, 1]])
print(data_df.iloc[0:2, 0:3])
print(data_df.iloc[:])
print(data_df.iloc[:, :])
print('맨 마지막 칼럼 데이터: \n', data_df.iloc[:, -1])
print('맨 마지막 칼럼을 제외한 모든 데이터: \n', data_df.iloc[:, : -1])
````

````python
print(data_df.loc['one', 'Name'])

# loc[]에 슬라이싱을 적용하면 종료값 -1이 아니라 종료값까지 포함
print(data_df.loc['one':'two', 'Name'])
print(data_df.loc['three', 'Name'])
print(data_df.loc['one':'two', ['Name', 'Year']])
print(data_df.loc['one':'three', 'Name':'Gender'])
print(data_df.loc[:])
print(data_df.loc[data_df.Year >= 2014])
````

#### 결측치 처리

````python
# 결측치 갯수 확인

df.isna().sum()
````

````python
# fillna로 결측치 대체

df['Cabin'] = df['Cabin'].fillna('C000')
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna('S')

df.isna().sum()
````

#### apply lambda

````python
# Name 칼럼의 문자열 개수 구하기

df['Name_len'] = df['Name'].apply(lambda x: len(x))
df[['Name', 'Name_len']].head(3)
````

````python
# 나이가 15세 이하면 Child, 그렇지 않으면 Adult로 구분

df['Child_Adult'] = df['Age'].apply(lambda x: 'Child' if x <= 15 else 'Adult')
df[['Age', 'Child_Adult']].head(10)
````

````python
# 나이가 15세 이하면 Child, 15 ~ 60세 사이는 Adult, 61세 이상은 Elderly로 분류

df['Age_cat'] = df['Age'].apply(lambda x: 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
df['Age_cat'].value_counts()
````

````python
# 5살 이하는 Baby, 12살 이하는 Child, 18살 이하는 Teenage, 25살 이하는 Student, 
# 35살 이하는 Young Adult, 60세 이하는 Adult, 그 이상은 Elderly로 분류

def get_category(age):
    cat = ''
    if age <= 5:
        cat = 'Baby'
    elif age <= 12:
        cat = 'Child'
    elif age <= 18:
        cat = 'Teenager'
    elif age <= 25:
        cat = 'Student'
    elif age <= 35:
        cat = 'Young Adult'
    elif age <= 60:
        cat = 'Adult'
    else:
        cat = 'Elderly'
    return cat

df['Age_cat'] = df['Age'].apply(lambda x: get_category(x))
df[['Age', 'Age_cat']].head()
````
